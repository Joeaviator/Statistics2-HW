```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Bestsellers with categories.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
Price_var=df['Price'].var()
print("Price variance: ",Price_var)
```

```python
Price_std=df['Price'].std()
print("Price STD: ",Price_std)
```

```python
User_var=df['User Rating'].var()
print("UserRating variance: ",User_var)
```

```python
User_std=df['User Rating'].std()
print("UserRating std: ",User_std)
```

```python
sns.histplot(data=df, x='Price', hue='Genre', palette='rainbow', bins=6)
plt.title("Price Distribution by Genre")
```

```python
sns.histplot(data=df, x='User Rating', bins=6)
plt.title("User Rating Distribution")
```
